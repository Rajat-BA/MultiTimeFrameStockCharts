```python
# Libraries
import pandas as pd
import plotly.graph_objects as go
import yfinance as yf

# Inputs
interval = input('Timeframe B=Days, W=week, M=month:')
stock_symbol = input('Please enter stock symbol on yahoo finance:')

# Stock data
df = yf.Ticker(stock_symbol).history(period = 'max')

# Resample intervals
df = df.resample(interval).agg({
    'Open':'first',
    'High':'max',
    'Low':'min',
    'Close':'last'
    })
# Drop null values
df.dropna(inplace = True)

# Create figure
fig = go.Figure()

# Plot candlestick chart
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Open'],
                             high=df['High'],
                             low=df['Low'],
                             close=df['Close']
                             ))

# Chart size
fig.update_layout(
    autosize=False,
    width= 1400,
    height= 700
    )

# Range buttons
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
                ])),
        rangeslider=dict(
            visible=True),
        type="date"
        ))
# Y-axis range
fig.update_layout(
      title = stock_symbol.upper() + ' Stock Chart',
      yaxis=dict(
      autorange = True,
      fixedrange= False
      ))
fig.show()
```
